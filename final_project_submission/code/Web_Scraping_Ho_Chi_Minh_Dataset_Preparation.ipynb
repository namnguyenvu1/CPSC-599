```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Ho_Chi_Minh')
soup = bs.BeautifulSoup(source, 'lxml')

text = ""
for paragraph in soup.find_all('p'):
  text += paragraph.text

# text = re.sub(r'\[[0-9]*\]',' ', text)
# text = re.sub(r'\[[a-f]*\]',' ', text)
# text = re.sub(r'\s+', ' ', text)
# text = text.lower()
# text = re.sub(r'\b\d{1,3}\b', ' ', text) # originally: text = re.sub(r'\d', ' ', text)
# text = re.sub(r'\s+', ' ', text)

# text = re.sub(r'\[[a-z]*\]',' ', text)  # This will remove [a], [b], etc.
# text = re.sub(r'\s+', ' ', text)
# # text = text.lower()
# text = re.sub(r'\b\d{1,3}\b', ' ', text)  # This will remove numbers that are not years.
# text = re.sub(r'\s+', ' ', text)
# text = re.sub(r'\[[ ]*\]',' ', text)
# text = re.sub(r'\s+', ' ', text)


text = re.sub(r'\[[a-z]*\]',' ', text)  # This will remove [a], [b], etc.
text = re.sub(r'\s+', ' ', text)
# text = text.lower()
text = re.sub(r'\[\d+\]', ' ', text) # This will remove [1], [2], etc.
text = re.sub(r'\s+', ' ', text)
text = re.sub(r'\[[ ]*\]',' ', text) # This will remove empty space and empty string, etc.
text = re.sub(r'\s+', ' ', text)
text = re.sub(r'\[.*?\]', ' ', text) # This will remove [citation needed], etc.
text = re.sub(r'\s+', ' ', text)
text = re.sub(r':\s*\d+(-\d+)?', ' ', text) # This will remove :51 or :126-128, etc.
text = re.sub(r'\s+', ' ', text)
```

```python
from transformers import BartForConditionalGeneration, BartTokenizer

# Load the model and tokenizer
model_name = "voidful/context-only-question-generator"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

# Define the contexts
contexts = sentences

# Initialize an empty list to store the questions
all_questions = []

i = 0
# Generate questions for each context
for context in contexts:
    # Tokenize the context
    inputs = tokenizer(context, return_tensors="pt", max_length=1024, truncation=True)

    print(i)
    # Generate questions
    outputs = model.generate(input_ids=inputs.input_ids)
    i = i + 1

    # Decode and append the questions to the list
    questions = tokenizer.batch_decode(outputs, skip_special_tokens=True)
    all_questions.append(questions)

# Print the generated questions
for i, context_questions in enumerate(all_questions):
    print(f"Questions for context {i+1}:")
    for j, question in enumerate(context_questions):
        print(f"{j+1}. {question}")
    print()
```

```python
# Print the generated questions
for i, context_questions in enumerate(all_questions):
    print(f"Questions for context {i+1}:")
    for j, question in enumerate(context_questions):
        print(f"{j+1}. {question}")
        df.Question[i] = question
    print()
```
