```python
# Import library
from sklearn.datasets import fetch_20newsgroups
from gensim.models import LdaModel
from gensim import corpora
```

```python
# Fetch the dataset
newsgroups = fetch_20newsgroups(remove=('headers', 'footers', 'quotes'))
```

```python
# Preprocess the data
import nltk
nltk.download('stopwords')

from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop = set(stopwords.words('english'))
lemma = WordNetLemmatizer()

def preprocess(text):
    tokens = word_tokenize(text)
    tagged = pos_tag(tokens)
    tokenized = [token.lower() for token, pos in tagged if token.isalpha() and pos != "IN" and token.lower() not in stop]
    normalized = [lemma.lemmatize(word) for word in tokenized]
    return normalized
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Apply preprocessing to all documents
processed_newsgroups = [preprocess(newsgroup) for newsgroup in newsgroups.data]
```

```python
# Create the dictionary and corpus
dictionary = corpora.Dictionary(processed_newsgroups)
corpus = [dictionary.doc2bow(newsgroup) for newsgroup in processed_newsgroups]
```

```python
# Train the LDA model and print top 10 topics
# lda_model = LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)
lda_model = LdaModel(corpus=corpus,
                     id2word=dictionary,
                     num_topics=10,
                     random_state=100,
                     update_every=1,
                     chunksize=100,
                     passes=10,
                     alpha='auto',
                     per_word_topics=True)
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

```python
from datasets import load_dataset
import pandas as pd
```

```python
dataset = load_dataset('MuskumPillerum/General-Knowledge')
```

```python
df_Hugging_Face = pd.DataFrame(dataset['train'])
# df_Hugging_Face.to_csv('General-Knowledge.csv', index=False)
```

```python
df_HCM = pd.read_csv("/content/cpsc_599_27_project_dataset.csv")
```

```python
# Add a new column 'Label' to both dataframes
df_Hugging_Face['Label'] = 'Not Related'
df_HCM['Label'] = 'Related'

# Merge the two dataframes and keep only the 'Question' and 'Label' columns
df = pd.concat([df_Hugging_Face, df_HCM])[['Question', 'Label']]
```

```python
from sklearn.model_selection import train_test_split

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df['Question'], df['Label'], test_size=0.2, stratify=df['Label'], random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

# Create a pipeline with a TF-IDF vectorizer and a logistic regression classifier
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LogisticRegression(class_weight='balanced'))
])

# Train the model
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

# Predict the labels for the test set
y_pred = model.predict(X_test)

# Print the classification report
print(classification_report(y_test, y_pred))
```

```python
test_pred = model.predict(["What was the subject of the article written by Hồ Chí Minh in May 1922, and who was implored to take action?"])
print(test_pred)
```
