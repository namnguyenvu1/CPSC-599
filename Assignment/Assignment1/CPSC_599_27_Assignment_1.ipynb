```python
# Remove digit
import re
text_remove_digit = []
for i in range(0, len(texts)):
  text_remove_digit.append(re.sub(r'\d+','', texts[i]))

# Then lower case
text_lower = []
for i in range(0, len(text_remove_digit)):
  text_lower.append(text_remove_digit[i].lower())
```

### Exercise 1B part 1 and 2

```python
# Exercise 1B (part 1 and 2 - tokenize using NLTK)
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize

# Tokenize sentences
sentences = []
for i in range(0, len(text_lower)):
  sentences.append(sent_tokenize(text_lower[i]))

# Tokenize words
words = []
for i in range(0, len(text_lower)):
  words.append(word_tokenize(text_lower[i]))
```

```python
# Print output
print(sentences[0])
print(words[0])
```

```python
%pip install spacy
!python3 -m spacy download en_core_web_sm
import spacy
nlp = spacy.load("en_core_web_sm")
nlp
```

```python
# Exercise 1B (part 1 and 2 - tokenize using SpaCy)
# Tokenize sentences (spicy_sentences is an array contain 10 text files after being sentence tokenized)
spicy_sentences = []
for i in range(0, len(text_lower)):
  spicy_sentence = nlp(text_lower[i])
  spicy_sentence_array = []
  for sent in spicy_sentence.sents:
    spicy_sentence_array.append(sent)
  spicy_sentences.append(spicy_sentence_array)

# Tokenize words (spicy_words is an array contain 10 text files after being words tokenized)
spicy_words = []
for i in range(0, len(spicy_sentences)):
  text_i = spicy_sentences[i] # text_i contain text_i.txt
  spicy_word_array = []
  for j in range(0, len(text_i)):
    tokens_spacy = [token.text for token in spicy_sentences[i][j]] # tokenize sentence j in text_i.txt into words
    spicy_word_array.append(tokens_spacy)
  spicy_words.append(spicy_word_array)
```

### Exercise 1B part 3

The difference between stemming and lemmatization is that stemming will simply remove the suffix of the words, while lemmatization will change the words to their simple form

For instance, after we apply stemming to the word "computers", we will get "comput" but if we apply lemmatization to "computers", we will get "computer" as output

As a consequence, applying lemmatization will make the words still retain the readable form of the words while stemming will simply remove the suffix of the words


-------------------------
#### How does NLTK and SpaCy differ in stemming and lemmatization?

In terms of stemming and lemmatization of NLTK and SpaCy, SpaCy don't support function for stemming and only have function for lemmatization. With NLTK, it supports both stemming and lemmatization.

With lemmatization, it seems like SpaCy can lemmatize words to their base form better than NLTK. For instance, the word "are" will be lemmatized to "are" using NLTK but will be lemmatized to "be" using SpaCy. This might be beneficial for the overall performance.

```python
# Stemming with NLTK
nltk.download('wordnet')
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

sentence = "These are super-computers."

porter = PorterStemmer()
print("Stemming with NLTK:")
for word in word_tokenize(sentence):
  print(porter.stem(word))

# Lemmatization with NLTK
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print("\nLemmatization with NLTK:")
for word in word_tokenize(sentence):
  print(lemmatizer.lemmatize(word))

# Lemmatization with SpaCy
doc = nlp("These are super-computers.")
print("\nLemmatization with SpaCy:")
for token in doc:
    print(token.lemma_)
```

```python
# Exercise 1C
from collections import defaultdict

# Set of words in each document
word_sets = [set(doc) for doc in words]
print(len(word_sets[0]))

doc_count = defaultdict(int)
for word_set in word_sets:
    for word in word_set:
        doc_count[word] += 1

threshold = 0.8 * len(words) # words here means the array of 10 input texts (it has length = 10)
common_words = [word for word, count in doc_count.items() if count >= threshold]
# List of common words
print("The list of common words are: ")
print(common_words)
```

```python
# Add stop words
nltk.download('stopwords')
nltk_stopwords = nltk.corpus.stopwords.words('english')
# Number of stop words before adding stop words
print(f"Number of stop words before adding stop words: {len(nltk_stopwords)}")

for i in common_words:
    if (i not in nltk_stopwords):
      nltk_stopwords.append(i)
# Number of stop words after adding stop words
print(f"Number of stop words after adding stop words: {len(nltk_stopwords)}")
```

```python
# Remove stopwords from text (text is store in words[i] with i from 0 to 9, 0 correspond to text_1.txt)
# filtered_words = [word for word in words if word not in nltk_stopwords]
print(len(words))
filtered_words = []
for i in range(len(words)):
  filtered_words.append([word for word in words[i] if word not in nltk_stopwords])

print(len(filtered_words))
print(filtered_words[0])
```
